```python
import pandas as pd

# CSV 파일 불러오기
file_path = '/content/drive/MyDrive/Project/test/test_labels.csv'
df = pd.read_csv(file_path)

# 'file_name;label' 열을 'file_name'과 'label'로 분리
df[['file_name', 'label']] = df['file_name;label'].str.split(';', expand=True)
df = df.drop(columns=['file_name;label'])
# 'file_name' 열을 3자리로 맞추기
df['file_name'] = df['file_name'].astype(str).str.zfill(3)
# 'label' 열에서 'x'를 1로, 'y'를 0으로 변환
df['label'] = df['label'].map({'generated': 1, 'real': 0})

# 이미지 경로 수정
df['file_name'] = '/content/drive/MyDrive/Project/test/test/' + df['file_name'] + '.jpg'
df.rename(columns={'file_name': 'image'}, inplace=True)
print(df.head())
print(df.columns)

# 수정된 데이터프레임 저장
new_file_path = '/content/drive/MyDrive/Project/test/test_binary.csv'
df.to_csv(new_file_path, index=False)
```
