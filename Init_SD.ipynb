```python
import os
import random

# 이미지가 있는 디렉토리 경로 설정
directory_path = '/content/drive/MyDrive/ArtificialIntelligence/Project/train/real'

# 디렉토리 내의 모든 파일 목록 가져오기
all_files = [os.path.join(directory_path, f) for f in os.listdir(directory_path) if os.path.isfile(os.path.join(directory_path, f))]

# 파일 목록에서 무작위로 5000개 선택
selected_files = random.sample(all_files, 50) if len(all_files) > 50 else all_files
```

```python
from diffusers import StableDiffusionPipeline
from PIL import Image
import torch


# Stable Diffusion 모델 파이프라인을 초기화. 이제 `variant`를 사용합니다.
pipe = StableDiffusionPipeline.from_pretrained(
    "CompVis/stable-diffusion-v1-4",
    variant="fp16",
    torch_dtype=torch.float16,
).to("cuda")

# Real face 이미지를 로드
input_image_path = '/content/drive/MyDrive/ArtificialIntelligence/Project/train/real/22480.jpg'
image = Image.open(input_image_path)

# 이미지 생성
prompt = 'A realistic human face image'
# of the person in the initial image

for i in range(3):
  generated_image = pipe(prompt=prompt, init_image=image, num_inference_steps=50, strength=0.8, guidance_scale=7.5).images[0]

  save_path = f"/content/drive/MyDrive/ArtificialIntelligence/Project/GeneratedImages/re_init_{i}.jpg"
  generated_image.save(save_path)
```

```python
# A photo of human face -> init_
# A portrait of human -> init
# A realistic face photo -> init__
# A realistic photo of human face -> init___
# 'A human face image' -> re_init
# 'A realistic human face image' -> re_init_
```
